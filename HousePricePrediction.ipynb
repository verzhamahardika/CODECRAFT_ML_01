## Data loading

### Subtask:
Load the training and testing datasets.


**Reasoning**:
Load the training and testing datasets into pandas DataFrames and display the first 5 rows of each.



```python
import pandas as pd

# Load the training dataset
df_train = pd.read_csv('datasets/train.csv')

# Load the testing dataset
df_test = pd.read_csv('datasets/test.csv')

# Display the first 5 rows of the training dataset
display(df_train.head())

# Display the first 5 rows of the testing dataset
display(df_test.head())
```

## Data exploration

Explore the training dataset to understand its structure, identify potential issues, and examine the distributions of key features.


**Reasoning**:
Explore the training dataset by examining its basic statistics, missing values, data types, and distributions of key features, as well as calculating and visualizing the correlation matrix of relevant features and the target variable.

## Data cleaning

### Subtask:
Clean the training and testing datasets.


**Reasoning**:
Impute missing values in 'LotFrontage', 'MasVnrArea', 'GarageYrBlt' and 'SalePrice', handle outliers in 'SalePrice' and 'GrLivArea', and apply the same cleaning steps to df_test.



```python
# Impute missing values
for df in [df_train, df_test]:
    df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
    df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
    df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())

# Handle outliers in 'SalePrice' and 'GrLivArea' for df_train
df_train = df_train[df_train['GrLivArea'] < 4500]
upper_limit = df_train['SalePrice'].quantile(0.99)
df_train = df_train[df_train['SalePrice'] < upper_limit]

# Verify data cleaning
print("Missing values in df_train:\n", df_train.isnull().sum().sort_values(ascending=False).head(10))
print("\nMissing values in df_test:\n", df_test.isnull().sum().sort_values(ascending=False).head(10))

# Display the cleaned data
display(df_train.head())
display(df_test.head())
```

## Feature engineering

### Subtask:
Prepare the features for the linear regression model.


**Reasoning**:
Select the relevant features ('GrLivArea', 'BedroomAbvGr', 'FullBath') from df_train and df_test to create X_train and X_test, and create y_train from 'SalePrice' in df_train.  Handle any missing values by imputation.



```python
import numpy as np

# Select features for X_train and X_test
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
X_train = df_train[features]
X_test = df_test[features]

# Create y_train
y_train = df_train['SalePrice']

# Impute missing values in X_train and X_test using the median
for col in features:
    X_train[col] = X_train[col].fillna(X_train[col].median())
    X_test[col] = X_test[col].fillna(X_test[col].median())

# Convert to NumPy arrays for model training
X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
```

## Data splitting

### Subtask:
Split the training data into training and validation sets.


**Reasoning**:
Split the training data into training and validation sets using train_test_split.



```python
from sklearn.model_selection import train_test_split

# Split data into training and validation sets
X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

## Model training

### Subtask:
Train a linear regression model.


**Reasoning**:
Train the linear regression model using the training data and make predictions on the validation set.



```python
from sklearn.linear_model import LinearRegression

# Instantiate the linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train_split, y_train_split)

# Make predictions on the validation set
y_pred = model.predict(X_val)
```

## Model evaluation

### Subtask:
Evaluate the linear regression model's performance on the validation set.


**Reasoning**:
Evaluate the linear regression model's performance using MSE, RMSE, and R-squared.



```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_val, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate R-squared (R2) score
r2 = r2_score(y_val, y_pred)

# Print the evaluation metrics
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2) Score: {r2}")
```

## Data preparation

### Subtask:
Prepare the test data for prediction using the same steps as the training data.


**Reasoning**:
Select the features from the test dataset, handle missing values using the same method as the training data, and convert the result to a NumPy array.



```python
# Select features for X_test
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
X_test_final = df_test[features]

# Impute missing values in X_test_final using the median from the training data
for col in features:
    X_test_final.loc[:, col] = X_test_final[col].fillna(df_train[col].median())

# Convert to NumPy array
X_test_final = X_test_final.values
```

## Model training

### Subtask:
Retrain the linear regression model using the entire training dataset.


**Reasoning**:
Retrain the linear regression model using the entire training dataset and make predictions on the test data.



```python
from sklearn.linear_model import LinearRegression

# Instantiate a new LinearRegression model
model_final = LinearRegression()

# Train the model using the entire training dataset
model_final.fit(X_train, y_train)

# Make predictions on the prepared test data
y_pred_final = model_final.predict(X_test_final)
```

## Model evaluation

### Subtask:

Load the sample submission file and replace the predictions with the new predictions. Then save the updated dataframe to a new csv file.



```python
import pandas as pd

# Load the sample submission file
submission_df = pd.read_csv('datasets/sample_submission.csv')

# Replace predictions
submission_df['SalePrice'] = y_pred_final

# Save to a new CSV file
submission_df.to_csv('submission.csv', index=False)

# Verify the submission file
print(submission_df.shape)
display(submission_df.head())
```
